```python
# This dictionary is keyed by author and contains a dictionary keyed by institution
# An institution's entry is the number of times the individual listed that institution
authoraff = {}
for a in authors:
    oldafflist = authors[a][11]
    newafflist= []
    for i in oldafflist:
        newi = affcodex2[i]
        newafflist.extend(newi)
    adict = {}
    for i in newafflist:
        temp = adict.get(i,0)
        temp += 1
        adict[i] = temp
    authoraff[a] = adict
```
